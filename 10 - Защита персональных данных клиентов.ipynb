```python
def error_inv(X):
    try: 
        np.linalg.inv(X)
        print('–ú–∞—Ç—Ä–∏—Ü–∞ –æ–±—Ä–∞—Ç–∏–º–∞')
    except:
        return error_inv(X)
```
